```python
import numpy as np
import matplotlib.pyplot as plt
import metodos_num as mn
```

`Ejemplo 1:`

$$f(x)=3x^2+9x$$

```python
p = lambda x: 3*x**2 + 9*x
interv = [-5, 5]

solRoot = mn.roo_Bis(p, interv)
solRootSv1 = mn.roo_Secv1(p, interv)
solRoot, solRootSv1
```

```python
xval = np.linspace(interv[0], interv[1], 1000)

plt.plot(xval, p(xval))
plt.plot(solRoot, [0]*len(solRoot), ls='', marker='o', markersize=5)
plt.hlines(y=0, xmin=-5, xmax=5, color='k', ls='--')
```
