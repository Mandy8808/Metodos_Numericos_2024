## Ejercicio 7 — Spline cuadrático

Construya un **spline cuadrático C¹** que interpole los datos anteriores.

Pasos sugeridos:

1. Suponga

\begin{align}
S_i(x) = a_i (x-x_i)^2 + b_i (x-x_i) + c_i
\end{align}

2. Imponga

\begin{align}
S_i(x_i) &= y_i\\

S_i(x_{i+1}) &= y_{i+1}\\

S_i'(x_{i+1}) &= S_{i+1}'(x_{i+1})
\end{align}

3. Use una condición adicional:

\begin{align}
S'(x_0) = 0
\end{align}
